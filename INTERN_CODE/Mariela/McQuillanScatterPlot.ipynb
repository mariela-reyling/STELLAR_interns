```python
%matplotlib inline

import matplotlib.pyplot as plt
import pandas as pd 
import numpy as np
```

```python
mq_data = pd.read_csv("data/McQuillan_data.csv")
```

mq_data

```python
mq_data

mass = mq_data["Mass"]
rp = mq_data["Prot"]
temp = mq_data["Teff"]
```

```python
plt.figure(figsize=(15,10))
plt.scatter(temp, rp, c = mass, s = 1)
plt.colorbar().set_label(label="Mass", size = 20)
plt.xlim(6500, 3500)
plt.xlabel("Temperature (K)", size = 30)
plt.ylabel("Rotation Period", size = 30)
plt.title("McQuillan 2014 Data", size = 30)
```
