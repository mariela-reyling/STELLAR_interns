```python
%matplotlib inline 
import matplotlib.pyplot as plt
import numpy as np 
import pandas as pd 
```

```python
path1 = open("Table_1_Periodic.txt","r")
path1.readline(10)
```

```python
table1= pd.read_csv("Table_1_Periodic.txt")
table1
```

```python
path2 = open("Table_2_Non_Periodic.txt","r")

table2= pd.read_csv("Table_2_Non_Periodic.txt")
table2
```

```python
len(table1)
```

```python
tables = pd.concat([table1,table2], axis=0, sort=True)
tables = tables.drop(columns=["Flag","Rper"])
tables
```

```python
tables
```

```python
valid = np.repeat(1,len(table1))
invalid = np.repeat(0,len(table2))  
classes = np.append(valid,invalid)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

```python
X = tables["Teff"]
```

```python
y = tables["Prot"]
```

```python
train_X, test_X, train_y, test_y = train_test_split(X, y)
model = DecisionTreeClassifier()
fit_model = model.fit(train_X,train_y)
predicted = fit_model.predict(test_X)
accuracy_score(test_y, predicted)
```
